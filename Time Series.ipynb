## About Dataset

This is a Data set for Stock Price of Netflix .
This Data set start from 2002 to 2022 .
It was collected from Yahoo Finance.
I took this dataset from Kaggle

https://www.kaggle.com/datasets/meetnagadia/netflix-stock-price-data-set-20

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

from datetime import datetime
```

```python

NFLX = pd.read_csv('C:\\Users\\saawa\\Downloads\\STA6857\\NFLX.csv') 
```

```python
sns.set(rc={'figure.figsize':(10,5)})

NFLX['Date'] = pd.to_datetime(NFLX['Date'])
NFLX = NFLX.set_index('Date')
NFLX.head()
```

```python
NFLX['Volume'].plot(label='x')

NFLX['Volume'].rolling(window=12).mean().plot(label='y')

sns.lineplot(x=NFLX.index, y=NFLX['Volume'], label='Volume')

sns.lineplot(x=NFLX.index, y=NFLX['Volume'].rolling(window=12).mean(), label='Averaged volume')
plt.title('Volume of stock versus time')
```

```python
NFLX['Open'].plot(label='x')

NFLX['Open'].rolling(window=12).mean().plot(label='y')

sns.lineplot(x=NFLX.index, y=NFLX['Open'], label='Open')

sns.lineplot(x=NFLX.index, y=NFLX['Open'].rolling(window=12).mean(), label='Averaged Open')
plt.title('Open of stock versus time')
```

```python
NFLX['Close'].plot(label='x')

NFLX['Close'].rolling(window=12).mean().plot(label='y')

sns.lineplot(x=NFLX.index, y=NFLX['Close'], label='Close')

sns.lineplot(x=NFLX.index, y=NFLX['Close'].rolling(window=12).mean(), label='Averaged Close')
plt.title('Close of stock versus time')
```

```python
NFLX['High'].plot(label='x')

NFLX['High'].rolling(window=12).mean().plot(label='y')

sns.lineplot(x=NFLX.index, y=NFLX['High'], label='High')

sns.lineplot(x=NFLX.index, y=NFLX['High'].rolling(window=12).mean(), label='Averaged High')
plt.title('High of stock versus time')
```

```python
NFLX.plot(y=['High','Low','Close','Open'],title='Netflix stock price')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,figsize=(15,10))
NFLX.groupby(NFLX.index.day).mean().plot(y=['High','Close','Low'],ax=ax1,xlabel='Day')
NFLX.groupby(NFLX.index.month).mean().plot(y=['High','Close','Low'],ax=ax2, xlabel='Month')
NFLX.groupby(NFLX.index.year).mean().plot(y=['High','Close','Low'], ax=ax3, xlabel='Year')
```

```python
# Example 1.1
NFLX.plot(xlabel='', ylabel='', marker='o', 
        legend=False)
plt.show()
```
